# Influencers

This notebook looks at conversations that were collected and sees how any of them were started by verified accounts, or accounts with more than 50,000 followers. It uses the `*_convs.json` files that were generated by the [Twitter Conversation Collection](Twitter%20Conversation%20Collection.ipynb) notebook.

```python
import os
import json
import pathlib

from twarc.client2 import Twarc2
from twarc import ensure_flattened

tweet_ids = set()
for conv_file in pathlib.Path('../data/json.terms.files').glob('*_convs.json'):
    for conv in json.load(conv_file.open())[0:100]:
        tweet_ids.add(conv['conversation_id'])

t = Twarc2(bearer_token=os.environ.get('BEARER_TOKEN'))
for resp in t.tweet_lookup(list(tweet_ids)):
    for tweet in ensure_flattened(resp):
        print(tweet['author']['username'], tweet['author']['public_metrics']['followers_count'])
```
